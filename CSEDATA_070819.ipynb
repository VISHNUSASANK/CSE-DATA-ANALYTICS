# Central Tendency and Variability

##### 1.mean(x) Mean of the numbers in vector x.
##### 2.median(x) Median of the numbers in vector x.
##### 3.var(x) Estimated variances of population from which the numbers in vector x are sampled.
##### 4.sd(x) Estimated standard deviation of population from which the numbers in vector x are sampled.

```r
x <- 1:10
cat(x)
cat('Mean of x is' ,mean(x))
cat( 'median of x is' ,median(x))
cat( 'Variance is' ,var(x))
cat( 'Standard deviation is' ,sd(x))
```

###### this problem can be solved by using new line function"\n" also.

```r
cat(x,'\n')
cat('Mean of x is' ,mean(x),'\n')
cat( 'median of x is' ,median(x),'\n')
cat( 'Variance is' ,var(x),'\n')
cat( 'Standard deviation is' ,sd(x)'\n')
```

```r
# add elements at the end
print('Elements in z before')
z
z <- c(z,'Bannureddy')
print('Elements in z after adding the last element')
z
```
